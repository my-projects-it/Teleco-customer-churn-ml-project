```python
# import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
df=pd.read_csv('Telco_Customer_Churn_Dataset  (1).csv') # load the dataset
print(df.head()) # display the first few rows of the dataset
```

```python
# separate features(X) and target(y)
X=df.drop('Churn', axis=1)
y=df['Churn']
# split the data
X_train, X_test, y_train, y_test=train_test_split(
    X, y, test_size=0.4, random_state=42
)
# display the shapes of the resulting datasets
print("Training Features Shape:", X_train.shape)
print("Testing Features Shape:", X_test.shape)
print("Training Features Shape:", y_train.shape)
print("Testing Features Shape:", y_test.shape)
# display a few rows of training and testing sets
print("Training features[X_train]:", X_train.head())
print("Testing features[X_test]:", X_test.head())
print("Training features[y_train]:", y_train.head())
print("Testing features[y_test]:", y_test.head())
```

# Conclusion:
# 1- The dataset was divided into training(80%) and testing(20%) sets using train_test_split to ensure a representative and unbiased split.
# 2- Stratification maintained consistent target class(Churn) distribution in both sets.
